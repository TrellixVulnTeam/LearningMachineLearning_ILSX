```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
import math


%matplotlib inline
```

```python
# make data
X,y = datasets.make_regression(n_samples=250, n_features=1, noise=5,random_state=0,bias=50)
print(X.shape)
print(y.shape)
#将完整的数据集切割成训练集和测试集
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=0)
#构造同样多行,1列的元素1矩阵
o_train = np.ones([X_train.shape[0],1],dtype=X_train.dtype)
o_test = np.ones([X_test.shape[0],1],dtype=X_test.dtype)
# print(o_train.shape)
# print(o_test.shape)
#将1矩阵与原矩阵按列合并(axis=1,按列合并)
X_train = np.concatenate((o_train,X_train),axis=1)
X_test = np.concatenate((o_test,X_test),axis=1)
print(X_train.shape)
print(y_train.shape)
```

```python
#show data
plt.figure(figsize=(10,7))
plt.ylim((-25,125)) #设置坐标范围
plt.xlim(-3,3)
plt.scatter(X_test[:,1],y_test,c='red',edgecolors='white')
plt.scatter(X_train[:,1],y_train,c='orange',edgecolors='white')

```

```python
#初始化w
THETA = np.zeros([2,1],dtype=np.float32)
#learning rate
lr_sto = 0.01
lr_bat = 0.0001
Epoch = 2
```

```python
def hypothesis(x,θ):
    h_x = np.matmul(θ.T,x)
    return h_x

def loss_function(h_x,y):
    J =((h_x-y)*(h_x-y)) **(1/2)
    return J

def update_parameters(THETA,h_x,x,y,lr):
    x = np.reshape(x,THETA.shape)
    #XTXW - XTy
    THETA = THETA - lr*(np.matmul(x.T,THETA)-y)*x
    return THETA

def batch_update_parameters(THETA,gradient,lr):
    THETA = THETA - lr*gradient
    return THETA
```

```python
# for plotting
p_x = np.linspace(-2.5, 2.5, 50)

THETA_list = []
i = 0

# stochastic gradient descent
# Epoch 整体数据重复执行几遍
for epoch in range(Epoch):
    for x, y in zip(X_train, y_train):
    # make hypothesis
        h_x = hypothesis(x, THETA)
    # compute loss
        loss = loss_function(h_x, y)
    # Excecute gradient descent to update parameters
        THETA = update_parameters(THETA, h_x, x, y, lr_sto)
        i += 1
        if i % 50 == 0:
            print("[{0}/{1}], loss: {2}".format(epoch+1, Epoch, loss[0]))
            plt.scatter(X_train[:, 1], y_train, c='purple', marker='o', edgecolors='white')
            plt.plot(p_x, THETA[1, 0] * p_x + THETA[0, 0])
            THETA_tmp = np.array(THETA[1, 0], THETA[0, 0])
            THETA_list.append(THETA_tmp)
            plt.ylim((-25, 125))
            plt.xlim((-3, 3))
            plt.show()
```

```python
def h(X, theta):
    return np.dot(X, theta)
  
def J(theta, X, Y):
    m = len(X)
    return np.sum(np.dot((h(X, theta)-Y).T , (h(X, theta)-Y)) / (2 * m))

def vector_squared_root(a):
    return math.sqrt(a.dot(a))
 
# 计算x,y坐标对应的高度值
def f(THETA1, THETA2, X_train, y_train, n):
    Loss_value = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            THETA = np.array([THETA1[i], THETA2[j]]).reshape(2,1)
      # compute loss
            loss = J(THETA, X_train, y_train)
            Loss_value[i, j] = loss
    return Loss_value
 
# 生成x,y的数据
n = 256
x = np.linspace(-50, 350, n)
y = np.linspace(-20, 50, n)
 
# 把x,y数据生成mesh网格状的数据，因为等高线的显示是在网格的基础上添加上高度值
X, Y = np.meshgrid(x, y)
# 填充等高线
plt.contourf(X, Y, f(x, y, X_train, y_train, n))
plt.plot(THETA_list)
# 显示图表
plt.show()
```
