```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
np.random.seed(1)
X = np.random.uniform(0,1,(40,1))
y = X**2 + np.random.normal(0,0.1,(40,1))
# y =x^2
X = np.sort(X,axis=0)
y= np.sort(y,axis=0)
plt.plot(X,y,'ro',label='original data')
```

```python
model_linear = LinearRegression()
model_linear.fit(X,y)
model_linear.score(X,y)
```

```python
#多项式回归(二次方拟合)
print(X.shape)
X_poly_right = np.array([X,X**2]).T.reshape(-1,2)
# X_poly_right2 = np.array([X,X**2]).reshape(-1,2)
print(X_poly_right.shape)
# print(X_poly_right2.shape)
# print(X_poly_right)
# print('--------------------------')
# print(X_poly_right2)
# print('--------------------------')
# print(X_poly_right-X_poly_right2)
model_right = LinearRegression()
model_right.fit(X_poly_right,y)
model_right.score(X_poly_right,y)
```

```python
#多项式回归(八次方拟合)
print(X.shape)
X_poly_eight = np.array([X,X**2,X**3,X**4,X**5,X**6,X**7,X**8]).T.reshape(-1,8)
print(X_poly_eight.shape)
model_eight = LinearRegression()
model_eight.fit(X_poly_eight,y)
model_eight.score(X_poly_eight,y)
```

```python
#画图显示
plot_predict = np.arange(-1,2,0.01).reshape(-1,1)
plot_predict_right = np.array([plot_predict,plot_predict**2]).T.reshape(-1,2)
plot_predict_eight = np.array([plot_predict,plot_predict**2,plot_predict**3,plot_predict**4,plot_predict**5,plot_predict**6,plot_predict**7,plot_predict**8]).T.reshape(-1,8)

y_plot = model_linear.predict(plot_predict)
y_plot_right = model_right.predict(plot_predict_right)
y_plot_eight = model_eight.predict(plot_predict_eight)

plt.figure(figsize=(18,7))
plt.ylim((-1,2))
plt.scatter(X,y,c='r',alpha=.5)
plt.plot(plot_predict,y_plot)
```

```python
plt.figure(figsize=(18,7))
plt.ylim((-1,2))
plt.plot(X,y,'ro',plot_predict,y_plot_right)  #注意这里x还是那个x, 仅仅是y改成二次方的预测值了
```
