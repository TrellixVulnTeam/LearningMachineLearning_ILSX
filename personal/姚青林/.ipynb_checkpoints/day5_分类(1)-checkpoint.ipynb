```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100,        # 样本个数
                           n_features=1,          # 特征个数
                           n_informative=1,        # 有效特征个数
                           n_redundant=0,          # 冗余特征个数（有效特征的随机组合）
                           n_repeated=0,           # 重复特征个数（有效特征和冗余特征的随机组合）
                           n_classes=2,            # 样本类别
                           n_clusters_per_class=1, # 簇的个数
                           random_state=0)
```

```python
X, y
```

```python
y[y==0] = -1
```

```python
y
```

```python
plt.scatter(X[:, 0], y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
b = model.intercept_
w = model.coef_[0]

w, b
```

```python
x0 = np.min(X[:,0])
x1 = np.max(X[:,0])
y0 = w * x0 + b
y1 = w * x1 + b

plt.plot([x0, x1], [y0, y1], c = 'r')
plt.scatter(X[:, 0], y)
plt.show()
```

```python
X, y = make_classification(n_samples=100,        # 样本个数
                           n_features=2,          # 特征个数
                           n_informative=1,        # 有效特征个数
                           n_redundant=0,          # 冗余特征个数（有效特征的随机组合）
                           n_repeated=0,           # 重复特征个数（有效特征和冗余特征的随机组合）
                           n_classes=2,            # 样本类别
                           n_clusters_per_class=1, # 簇的个数
                           random_state=0)
y[y==0] = -1
y
```

```python
model = LinearRegression()
model.fit(X, y)
b = model.intercept_
w1 = model.coef_[0]
w2 = model.coef_[1]

w1, w2, b
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

X_1 = X[y==1]
X_n1 = X[y==-1]

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_1[: , 0], X_1[: ,1], y[y==1])
ax.scatter(X_n1[: , 0], X_n1[: ,1], y[y==-1], c='r')

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X * w1 + Y * w2 + b

# 具体函数方法可用 help(function) 查看，如：help(ax.plot_surface)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

```
