# 正态分布

```python

#PDF 概率密度函数
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100))/np.max(stats.norm.pdf(np.linspace(-3,3,100))) ,
        )
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100))/np.max(stats.norm.pdf(np.linspace(-3,3,100))),
        alpha = .15,
        )
#CDF 累积概率密度函数
plt.plot(np.linspace(-4,4,100),
        stats.norm.cdf(np.linspace(-4,4,100))
        )

#LEGEND
plt.text(x=-1.5, y=.7, s='pdf (normed)',rotation=65, alpha=.75, weight='bold', color='#008fd5')
plt.text(x=.4, y=.5, s='cdf', rotation=55,alpha=.75,weight='bold',color = '#fc4f30')

#TICKS
plt.tick_params(axis='both',which='major', labelsize=18)
plt.axhline(y=0,color='black',linewidth=1.3,alpha=.7)

#TITLE
# plt.text(x=-5,y=1.2,s='Normal Distribution Overview',fontsize=26,weight='bold',alpha=.75)
# plt.title('正态分布图',fontproperties=myfont)
plt.text(x=-5,y=1.2,s='正态分布图',fontsize=32,weight='bold',alpha=.75)
```

```python
#PDF MU=0
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100)),
        )
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100)),
        alpha=.15)
#PDF MU=2
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),loc=2),
        )
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),loc=2),
        alpha=.15)
#PDF MU=-2
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),loc=-2),
        )
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),loc=-2),
        alpha=.15)



plt.tick_params(axis='both',which='major', labelsize=18)
plt.axhline(y=0,color='black',linewidth=1.3,alpha=.7)
plt.text(x=-5,y=0.45,s='正态分布图',fontsize=32,weight='bold',alpha=.75)
```

```python
#标准差
#PDF sigma =1
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),scale=1) ,
        )
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),scale=1),
        alpha = .15,
        )

#PDF sigma =2
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),scale=2) ,
        alpha = .35)
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),scale=2),
        alpha = .15,
        )

#PDF sigma =0.5
plt.plot(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),scale=0.5) ,
        )
plt.fill_between(np.linspace(-4,4,100),
        stats.norm.pdf(np.linspace(-4,4,100),scale=0.5),
        alpha = .7,
        )
```

```python
#概率密度函数(probability densisty function)
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (14,7)
plt.figure(dpi=100)
x=-1
y=2
print('pdf(x)={}\npdf(y)={}'.format(norm.pdf(x),norm.pdf(y)))
x_s = np.linspace(-3,3,100)
y_s = norm.pdf(x_s)
y_s2 = norm.cdf(x_s)
plt.scatter(x_s,y_s)
#累积概率密度函数(Cumlative probability density)
plt.scatter(x_s,y_s2)
print(norm.cdf(0.3)-norm.cdf(-0.3))
#计算取值(-0.3,0.3)的概率
```

```python
plt.figure(dpi=100)
mu_real = 10
sigma_real =2
np.random
sample = stats.norm.rvs(loc=mu_real,scale = sigma_real,size=1000)
mu_est = np.mean(sample)
sigma_est = np.std(sample)
print('mu估计值:{}\n sigma估计值:{}'.format(mu_est,sigma_est))

####plotting######
#sample distribution
plt.hist(sample,bins=100,normed=True,alpha=.25)
#true curve
plt.plot(np.linspace(2,18,1000),norm.pdf(np.linspace(2,18,1000),loc=mu_real,scale=sigma_real))
#estimated curve
plt.plot(np.linspace(2,18,1000),norm.pdf(np.linspace(2,18,1000),loc=mu_est,scale=sigma_est))

#ticks
plt.tick_params(axis='both',which='major',labelsize=18)
plt.axhline(y=0,color='black',linewidth=1.3,alpha=0.7)
```

# 二项式分布

```python
%matplotlib inline
style.use('fivethirtyeight')
# myfont = matplotlib.font_manager.FontProperties(fname="Light.ttc") 
# #解决显示中文问题, 后面需要中文的地方加上fontproperties=myfont
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False
#解决中文显示尝试二
plt.rcParams['figure.figsize']=(14,7)
plt.figure(dpi=100)

#PDF 概率密度函数
plt.bar(x=np.arange(20),
        height=(stats.binom.pmf(np.arange(20),p=.5,n=20)),
        width=.75,
        alpha=.75
        )
#CDF 累积概率密度函数
plt.plot(np.arange(20),
        stats.binom.cdf(np.arange(20),p=.5,n=20),
        color='#fc4f30'
        )


#LEGEND
plt.text(x=4.5, y=.7, s='pmf (normed)', alpha=.75, weight='bold', color='#008fd5')
plt.text(x=14.5, y=.9, s='cdf',alpha=.75,weight='bold',color = '#fc4f30')

#TICKS
plt.xticks(range(21)[::2])
#TITLE
# plt.text(x=-5,y=1.2,s='Normal Distribution Overview',fontsize=26,weight='bold',alpha=.75)
# plt.title('正态分布图',fontproperties=myfont)
plt.text(x=-5,y=1.2,s='正态分布图',fontsize=32,weight='bold',alpha=.75)
```
