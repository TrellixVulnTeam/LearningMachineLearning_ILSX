```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy import linalg
from sklearn.datasets import make_regression
```

```python
def simple_regression(X,Y):
    k = (((X-X.mean())*(Y-Y.mean())).sum())/np.dot(X-X,mean(),X-X.mean())
    b = Y.mean() - k*X.mean()
    Y_theory = k*X+b
    return k,b,Y_theory
```

```python
data_1 = make_regression(1000,1,1,1,0.1,1,noise=.2,random_state=145)
```

```python
X_own = data_1[0].reshape(1,-1)[0]
Y_own = data_1[1]
Ytheory = 1
```

```python
lin = linear_model.LinearRegression()
lin.fit(data_1[0],data_1[1])
lin.coef_,lin.intercept_
```

```python
dgs,axes1 = plt.subplots(1,2,dpi=140,figsize=(14,8))
axes1[0].scatter(X_own,Y_own,c='m',marker='*')
axes1[0].plot(X_own,Ytheory,'-c')
axes1[0].plot(X_own,(Y_own-Ytheory),'Dk')
axes1[1].plot(X_own,(Y_own-Ytheory),'Dk')
```
